#  Olá, seja bem-vindo ao nosso projeto de redução de dimensionalidade de imagens utilizando Python.
### Bom, você deve estar se perguntando qual a necessidade de reduzir a dimensão das imagens, sendo que atualmente estamos vivendo um mundo onde os computadores nunca foram tão robustos e capazes em termos de armazenamento e processamento de imagens, porém temos que levar em consideração que as câmeras e aparelhos que registram essas imagens também evoluíram com o tempo e hoje um simples vídeo ou imagem pode ocupar milhares de bytes de memória, sendo assim, se torna necessário diminuir esse tamanho na memória para que possamos utilizá-los em aplicações de aprendizado de máquina..    

```python
from PIL import Image

caminho_imagem = "papai_noel.jpg"
imagem_original = Image.open(caminho_imagem)
```

Primeiramente iremos carregar nosso "module" **Image** do "Package" **PIL**, após isso iremos criar duas variáveis, uma para especificar o caminho da imagem e outra para se referir a essa imagem para que possamos utilizá-la no nosso código.

```python
def converter_para_cinza(imagem):
  largura, altura = imagem.size
  imagem_cinza = Image.new("L", (largura, altura))
  for y in range(altura):
      for x in range(largura):
          r, g, b = imagem.getpixel((x, y))
          valor_cinza = int(0.299 * r + 0.587 * g + 0.114 * b)
          imagem_cinza.putpixel((x, y), valor_cinza)

  return imagem_cinza
```

Em seguida iremos definir uma função que pega os pixels dessa imagem e multiplica os valores RGB com valores pré-definidos que irão definir qual o tom de cinza aquele respectivo pixel vai ter após a execução desse calculo a imagem é armazenada em outra variável que iremos utilizar mais pra frente.

```python

def binarizar_imagem(imagem_cinza, limiar=127):
  largura, altura = imagem_cinza.size
  imagem_binaria = Image.new("1", (largura, altura))

  for y in range(altura):
      for x in range(largura):
          valor_cinza = imagem_cinza.getpixel((x, y))
          valor_binario = 1 if valor_cinza > limiar else 0
          imagem_binaria.putpixel((x, y), valor_binario)

  return imagem_binaria
```

Agora definimos uma função que utiliza a imagem resultado da função anterior para transformar essa imagem em uma imagem binarizada, que consiste basicamente em uma imagem com apenas dois valores ou preto ou branco.

```python
from IPython.display import display
from PIL import Image

imagem_cinza = converter_para_cinza(imagem_original)
imagem_binaria = binarizar_imagem(imagem_cinza)

def exibir_imagem(imagem):
  size_img = 200
  return display(imagem.resize((size_img,size_img)))

print("Imagem Original:")
exibir_imagem(imagem_original)

print("Imagem em Tons de Cinza:")
exibir_imagem(imagem_cinza)

print("Imagem Binarizada:")
exibir_imagem(imagem_binaria)
```
